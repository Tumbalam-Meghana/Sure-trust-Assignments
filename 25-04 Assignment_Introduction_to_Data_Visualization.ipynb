# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

### Expected output:


```python
# Your code here
bar_width = 0.4
months = df['month_number']
facecream_sal = df['facecream']
facewash_sal = df['facewash']
r1 = months - bar_width/2
r2 = months + bar_width/2
plt.bar(r1, facewash_sal, width=bar_width, label='Face Wash', color='blue')
plt.bar(r2, facecream_sal, width=bar_width, label='Face Cream', color='orange')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Product Sales Month Wise')
plt.legend()
plt.show()
```
