```python
import pandas as pd
data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'IT', 'HR'],
        'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
        'Salary': [50000, 70000, 45000, 80000, 60000, 62000, 72000, 48000],
        'Experience': [5, 7, 4, 10, 8, 6, 9, 3]}
```

```python
df=pd.DataFrame(data)
df
```

```python
#Find the total and average salary for each department.
df_sal=df.groupby(by="Department",as_index=False)["Salary"].mean()
df_sal
```

```python
#max
df.groupby(by="Department",as_index=False)["Salary"].max()
```

```python
#Add a new column that shows the average salary of each department and assign it to each employee.
df_mod=pd.merge(df,df_sal,how="left",on="Department")
df_mod
```

```python
df_mod.rename(columns={"Salary_x":"Salary",
                       "Salary_y":"Average_Salary"},inplace=True)
df_mod
```

```python
#By using Apply function Add a new column that shows the average salary of each department and assign it to each employee.
def get_avg_salary(dept):
    df_avg_Sal=df.groupby(by="Department",as_index=False)["Salary"].mean()
    avg_salary_value=df_avg_Sal[df_avg_Sal["Department"]==dept]["Salary"].iloc[0]
    return avg_salary_value
df["Avg_salary"]=df["Department"].apply(lambda value:get_avg_salary(value))
df
```

```python
"""Find the total salary of employees grouped by Experience range:

0-5 years: 'Junior'

6-10 years: 'Senior'"""
def get_category_name(experience_value):
    if experience_value<=5:
        return "Junior"
    elif experience_value>=6 and experience_value<=10:
        return "Senior"
df["employee_category"]=df["Experience"].apply(lambda value:get_category_name(value))
df
```

```python
import pandas as pd
marks_data = {'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
              'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'Math', 'Science'],
              'Marks': [85, 78, 92, 88, 76, 95, 89, 80]}
```

```python
df=pd.DataFrame(marks_data)
df
```

```python
#Find the average marks for each subject.

df_sub=df.groupby(by="Subject",as_index=False)["Marks"].mean()
df_sub
```

```python

#Find the highest and lowest marks in each subject.
max_marks = df.groupby('Subject', as_index=False)['Marks'].max().rename(columns={'Marks': 'Highest Marks'})
min_marks = df.groupby('Subject', as_index=False)['Marks'].min().rename(columns={'Marks': 'Lowest Marks'})
print(max_marks)
print(min_marks)
```

```python
sales_data = {'City': ['New York', 'New York', 'Paris', 'Paris', 'London', 'London', 'New York', 'Paris'],
              'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'A'],
              'Sales': [100, 200, 150, 250, 130, 220, 170, 180]}

df_sales = pd.DataFrame(sales_data)
```

```python
#Calculate the total and average sales per city.

total=df_sales.groupby('City')['Sales'].sum()
print(total,'\n')
avg=df_sales.groupby('City')['Sales'].mean()

print(avg)
```

```python
#Find out which city had the highest sales for each product.
highest_sales=df_sales.groupby('Product')['City'].max()
print(highest_sales)
```

```python
#Create a pivot table that shows total sales per city and product.
pivottable=df_sales.pivot_table(values='Sales',index='City',aggfunc='sum')
print(pivottable)
```
