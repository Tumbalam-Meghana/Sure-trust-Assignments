```python
data = df[['Age', 'Fare', 'Survived']].to_numpy()
print(data)
```

```python
#a)Remove rows with missing values in any of the selected columns.
df_clean = data[~np.isnan(data).any(axis=1)]  # remove rows with N
df_clean
```

```python
#For normalization we need to Extract age, fare, and survival columns
ages = df_clean[:, 0]
fares = df_clean[:, 1]
survived = df_clean[:, 2]
# b) Normalize age and fare using Z-score normalization
ages = df_clean[:, 0]
fares = df_clean[:, 1]
survived = df_clean[:, 2]
print(ages)
print(fares)
```

```python
mean_age_survived = np.mean(ages[survived == 1])
mean_fare_survived = np.mean(fares[survived == 1])
mean_age_not_survived = np.mean(ages[survived == 0])
mean_fare_not_survived = np.mean(fares[survived == 0])
print("Mean Age of the Survived:", mean_age_survived)
print("Mean Fare of the Survived:", mean_fare_survived)
print("Mean Age for the Not Survived:", mean_age_not_survived)
print("Mean Fare of the Not Survived:", mean_fare_not_survived)
```

```python
# d)BY using the np.where() for  classifying  the fare as 'Low' or 'High'
fare_mean = np.mean(fares)
fare_class = np.where(fares < fare_mean, "Low", "High")

# Print first 10 fare classifications as a sample
print("Fare Class First 10 of the classifications:", fare_class[:10])
```

```python
#Here is the code with explanations:

#Loading the Iris Dataset and Creating Petal Ratio

import pandas as pd

# Load the iris dataset from a CSV file into a DataFrame called df1
df1 = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\pythonprograms\iris_dataset.csv")

# Create a new column called 'petal_ratio' in the DataFrame df1, which is the ratio of 'Petal.Length' to 'Petal.Width'
df1['petal_ratio'] = df1['petal.length'] / df1['petal.width']


#Average Petal Ratio

# Calculate and print the mean of the 'petal_ratio' column in the DataFrame df1
print(df1['petal_ratio'].mean())


#Filtering Dataset: Sepal Width Above Mean

# Filter the DataFrame df1 to only include rows where 'Sepal.Width' is greater than the mean of 'Sepal.Width', and print the result
print(df1[df1['sepal.width'] > df1['sepal.width'].mean()])


#Splitting DataFrame: Petal Ratio < 2 and ≥ 2

# Split the DataFrame df1 into two subsets: subset1 where 'petal_ratio' is less than 2, and subset2 where 'petal_ratio' is greater than or equal to 2
subset1 = df1[df1['petal_ratio'] < 2]
subset2 = df1[df1['petal_ratio'] >= 2]

# Concatenate subset1 and subset2 vertically and print the result
print(pd.concat([subset1, subset2]))
```
